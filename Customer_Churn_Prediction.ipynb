```python
def plot_distribution(df, column_name):
  plt.figure(figsize=(12,4))
  plt.subplot(1,2,1)
  sns.histplot(df[column_name], kde=True)
  plt.title(f"Distribution of {column_name}")

  col_mean = df[column_name].mean()
  col_median = df[column_name].median()
  plt.axvline(col_mean, color='red', linestyle="--", label="Mean")
  plt.axvline(col_median, color='blue', linestyle="--", label="Median")

  plt.legend()

  plt.subplot(1,2,2)
  sns.boxplot(df[column_name])
  plt.title(f"Boxplot of {column_name}")
  plt.show()
```

```python
## Label Encoding

df['Churn'] = df['Churn'].replace({"Yes":1, "No":0})
```

```python
df.info()
```

```python
objects_columns = df.select_dtypes(include = "object").columns
```

```python
objects_columns
```

```python
from sklearn.preprocessing import LabelEncoder

encoders = {}

for column in objects_columns:
  label_encoder = LabelEncoder()
  df[column] = label_encoder.fit_transform(df[column])
  encoders[column] = label_encoder

```

```python
encoders
```

```python
import pickle
with open("encoder.pkl", "wb") as f:
  pickle.dump(encoders, f)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
##SMOTE Technique
## Using oversampling as data is balanced

from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
```

```python
x = df.drop(columns = ['Churn'])
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
y_train.value_counts()
```

```python
##SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
y_train_smote.value_counts()
```

```python
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42)
}
```

```python
hyperparameters = {
    # Random Forest Hyperparameters
    "Random Forest": {
        "n_estimators": [50, 100, 200],
        "max_depth": [5, 10, None]  # None means no max depth limit
    },

    # XGBoost Hyperparameters
    "XGBoost": {
        "learning_rate": [0.01, 0.1, 0.2],
        "max_depth": [3, 5, 7]
    }
}
```

```python
best_models = {}
for model_name, model in models.items():
  print(f"Tunning {model_name}....")
  grid = GridSearchCV(estimator=model, param_grid=hyperparameters[model_name], cv=3, scoring='accuracy')
  grid.fit(X_train_smote, y_train_smote)
  best_models[model_name] = grid.best_estimator_
  print(f"Best Parameters for {model_name} : {grid.best_params_}")
  print(f"Best Accuracy for {model_name} : {grid.best_score_: .2f}\n")

```

```python
with open("best_models.pkl", "wb") as f:
  pickle.dump(best_models['Random Forest'], f)
```

```python
with open("scaler.pkl", "wb") as f:
  pickle.dump(scaler, f)
```

```python
y_test_pred = best_models['Random Forest'].predict(X_test)
y_test_prob = best_models['Random Forest'].predict_proba(X_test)[:, 1]
```

```python
y_test_prob
```

```python
print(f"Accuracy: ", accuracy_score(y_test, y_test_pred))
print(f"ROC -AUC SCORE: ", roc_auc_score(y_test, y_test_pred))
print(f"Confusion Matrix: ", confusion_matrix(y_test, y_test_pred))
print(f"Classification report: ", classification_report(y_test, y_test_pred))

```

```python
with open("best_model.pkl", "rb") as f:
    loaded_model = pickle.load(f)
with open("encoder.pkl", "rb") as f:
    encoders = pickle.load(f)
with open("scaler.pkl", "rb") as f:
    scaler_data =pickle.load(f)
```

```python
def make_prediction(input_data):
    input_df = pd.DataFrame([input_data])
    for col, encoder in encoders.items():
        input_df[col] = encoder.transform(input_df[col])
    numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
    input_df[numerical_cols] = scaler_data.transform(input_df[numerical_cols])
    prediction = loaded_model.predict(input_df)[0]
    probability = loaded_model.predict_proba(input_df)[0][1]
    return "Churn"  if prediction == 1 else "No Churn", probability
```
